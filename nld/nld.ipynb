```python
import numpy as np
import matplotlib.pyplot as plt

# Define the original function F(x_n)
def F(x):
    return(1-2*np.abs(x-(1/2)))

# Define the recursive composition function
def compose_function(x, n):
    if n == 0:
        return x
    else:
        return F(compose_function(x, n - 1))

# Generate a range of x_n values
x_values = np.linspace(0, 1, 5)

# Choose the number of compositions (n)
n = 3  # Change this value as needed

# Calculate the corresponding values of the composed function
composed_values = compose_function(x_values, n)

# Create the plot
plt.plot(x_values, composed_values)
plt.xlabel('x_n')
plt.ylabel(f'Composed F(x_n) ({n} times)')
plt.title(f'Composed Function: F composed {n} times')
plt.grid(True)
plt.show()
```
